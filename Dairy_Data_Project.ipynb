```python
# This is the syntax to use a data set that we learned in class

import pandas as pd
dairy = pd.read_csv("Cottage.csv", low_memory = False)
dairy
```

```python
# Renaming column titles
dairy2 = dairy.rename(columns={"Dairy products: Per capita consumption, United States (pounds per person) 1": "Year", "Unnamed: 1": "Fluid_Bev_Milk", "Unnamed: 2": "American_Type_Cheese", "Unnamed: 3": "Non_American_Cheese", "Unnamed: 4": "Cottage", "Unnamed: 5": "Butter", "Unnamed: 6": "Dry_Whole_Milk", "Unnamed: 7": "Dry_Skim_Milk", "Unnamed: 8": "Dry_Buttermilk", "Unnamed: 9": "Dry_Whey_Protein", "Unnamed: 10": "Regular_Ice_Cream", "Unnamed: 11": "Low_Fat_Ice_Cream", "Unnamed: 12": "Frozen_Yogurt", "Unnamed: 13": "Sherbet", "Unnamed: 14": "Other_Frozen_Dairy", "Unnamed: 15": "Water_Juice_Ices", "Unnamed: 16": "Yogurt", "Unnamed: 17": "Evap_Cond_Milk_Whole_Canned", "Unnamed: 18": "Evap_Cond_Milk_Whole_Bulk", "Unnamed: 19": "Evap_Cond_Milk_Skim_Bulk_Canned", "Unnamed: 20": "All_Products_Milk_Fat_Basis", "Unnamed: 21": "All_Products_Skim_Solids_Basis"})


# Removing first 4 rows and last rows
dairy2 = dairy2.drop(labels=range(0, 4), axis=0)
dairy2 = dairy2.drop(labels=range(51, 59), axis=0)


# Removing Skim-Basis column because it is mostly NaN values
dairy2 = dairy2.drop(labels = ["All_Products_Skim_Solids_Basis"], axis = 1, inplace = False) 

# Filling in the NaN values in the rest of the data set
#converting values to float so we can take the mean
dairy2["Frozen_Yogurt"]=dairy2["Frozen_Yogurt"].astype(float)
print(dairy2["Frozen_Yogurt"].mean())
#filling the NaN values with the mean to maintain the integrity of the data
dairy2["Frozen_Yogurt"] = dairy2["Frozen_Yogurt"].fillna(1.75) #1.75 is the mean we calculated 
dairy2
```

```python
import matplotlib.pyplot as plt

# Class which takes in a column as a parameter, makes a table with that column and "Years", and then plots that column over time

class over_time:
    def __init__(self, column, type = "line", color = "blue"):
        self.column = column
        self.type = type
        self.color = color
#initializing the class with mandatory column parameter to be plotted against time, as well as optional parameters to change the graph type and color
    def make_table(self):
        global ot_table
        ot_table = dairy2.groupby("Year")[[self.column]].mean().round()
        return ot_table
#function within the class that makes a table with the year and column input
    def graph_ot(self):
        global ot_table
        return ot_table.plot(kind = self.type, color = self.color)
#function within the class that takes the table created by the table function and plots the graph

        
        
c1 = over_time("Fluid_Bev_Milk", "line", "sienna")
c1.make_table()
c1.graph_ot()

c2 = over_time("All_Products_Milk_Fat_Basis", "line", "wheat")
c2.make_table()
c2.graph_ot()

c3 = over_time("Yogurt", "bar", color = "yellowgreen")
c3.make_table()
c3.graph_ot()

#The downsides of this structure are the several steps needed to change the graph parameters and the way the graphs look, these parameters need to be added to the class in order to change them. However, with this format it makes graphing a lot easier, especially if you want graphs from the same dataset, and for the graphs to have a cohesive and predetermined look
```

```python
# Changing the data type of all columns from objects to floats for numeric analysis

dairy3 = dairy2.copy()
# Syntax: dairy3["Year"]=dairy3["Year"].astype(int)

for i in dairy3.columns:
    if i == "Year":
        dairy3[i]=dairy3[i].astype(int)
    else:
        dairy3[i]=dairy3[i].astype(float)

print(dairy3.dtypes)
```

```python
# This plots Milk consumption and All Dairy Products consumption on the same graph to see the comparison using the same y axis scale
comparison_ot = dairy3.plot(x="Year", y=["Fluid_Bev_Milk", "All_Products_Milk_Fat_Basis"], kind="line", rot=0)
```

```python
#Syntax to create a pie graph to visualize the different cheeses!

labels = "American_Type_Cheese","Non_American_Cheese","Cottage"

# sum returns the sum of every value from a single column
print(dairy3.sum("rows"))

# American: 576.7, Non_American: 723.1, Cottage: 142.8 

cheese = [576.7,723.1,142.8]

explode = (0,0,.5)

colors = "olivedrab","yellowgreen","lemonchiffon"

fig, ax1 = plt.subplots()
ax1.pie(cheese,colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow=True, startangle=120)
ax1.axis("equal")

plt.show()


# Pie chart showing the breakdown of all the frozen dairy products
labels1 = "Ice Cream","Lowfat Ice Cream","Frozen Yogurt","Sherbet","Other","Water Juice Ices"
frozen = [720.5,301.3,57.8,51.8,26.4,60.5]
colors1 = "cadetblue","powderblue","steelblue","midnightblue","cornflowerblue","lightsteelblue"

fig, ax2 = plt.subplots()
ax2.pie(frozen,colors = colors1, labels = labels1, shadow=True, startangle=100)
ax2.axis("equal")

plt.show()
```
